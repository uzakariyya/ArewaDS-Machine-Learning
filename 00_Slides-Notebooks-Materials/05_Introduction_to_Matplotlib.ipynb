# Data Visualization

> As data scientists, one of our most powerful tools is the ability to visualize data. A well-crafted chart or graph can reveal insights that raw numbers alone cannot. 

> As the saying goes, "A picture is worth a thousand words," and in data science, a good visualization can expose trends, outliers, patterns, and relationships that help us understand the story behind the data.



> Visualizations is sometimes called plots

## Types  

we will explore four key categories of data visualization, each serving a different purpose:


- Visualizing Quantities – Comparing magnitudes.

- Visualizing Distributions – Understanding spread and shape.

- Visualizing Proportions – Showing parts of a whole.

- Visualizing Relationships – Examining correlations and interactions.



## Plotting Libraries in Python

Python provides several powerful libraries for data visualization:

- **Matplotlib**: The core 2D plotting library in Python. Offers full control over plot appearance. Ideal for publication-quality figures.

- **Seaborn**: Built on Matplotlib. Simplifies the creation of attractive statistical graphics.

- **Plotly / Bokeh**: Interactive and web-friendly plotting libraries.

- **Altair**: Declarative approach for clean, statistical plots with minimal code.

### Why We Teach Matplotlib

- **Foundational**: Underpins many other libraries (e.g., Seaborn, pandas).

> Learning Matplotlib lays a strong foundation for mastering other plotting tools in Python.

## Figures and Subplots

Plots in matplotlib reside within a Figure object. You can create a new figure with `plt.figure:`



```python
fig = plt.figure()
```

#### Figure 

- Think of a figure `(plt.figure())` as a blank sheet of paper where you'll draw your plots.

- It is the top-level container that holds everything (all your charts, graphs, etc.).

- You can have multiple figures open at once, each in its own window.


```python
fig = plt.figure()          # Create a blank figure
ax1 = fig.add_subplot(1, 1, 1)  # Add the first subplot
```

#### Subplots (Individual Plots)
  
- Subplots are individual graphs/charts that you place on your figure.

- The `add_subplot()` method adds a single subplot to your figure.

- The numbers in (2, 2, 1) mean:

  - First number (2): Number of rows in your subplot grid
  
  - Second number (2): Number of columns in your subplot grid

  - Third number (1): Position of this specific subplot in the grid

- plt.figure has a number of options; `figsize` which will guarantee the figure has a certain size and aspect ratio if saved to disk.

- You can’t make a plot with a blank figure. You have to create one or more subplots using add_subplot:





```python
fig = plt.figure()
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
```

> These plot axis objects have various methods that create different types of plots, and it is preferred to use the axis methods over the top-level plotting functions like `plt.plot`. For example, we could make a line plot with the plot method (see Data visualization after a single plot):

#### Figure Size

In Matplotlib, `plt.figure(figsize=(width, height))` controls the size of your entire plot figure (the canvas that holds all your charts). 

- `figsize=(width, height)` accepts dimensions in inches. Example: figsize=(10, 5) 

- First number `(10) = width (left-to-right`), Second number `(5) = height (bottom-to-top)`

- Default size is `(6.4, 4.8)`, which might be too small for: Multiple subplots, Complex data and Publication-quality figures
 

#### Wide Plot (Good for time series):



```python
plt.figure(figsize=(10, 3))  # Wide and short
plt.plot([1, 2, 3], [4, 5, 6])
plt.title("Wide Plot")
plt.show()
```

Square Plot (Good for scatter plots):

```python
plt.figure(figsize=(6, 6))  # Perfect square
plt.scatter([1, 2, 3], [4, 5, 6])
plt.title("Square Plot")
plt.show()
```

Tall Plot (Good for stacked subplots):



```python
plt.figure(figsize=(5, 10))  # Tall and narrow
# Add 3 vertical subplots
for i in range(3):
    plt.subplot(3, 1, i+1)
    plt.plot([1, 2, 3], [i, i*2, i*3])
plt.tight_layout()
plt.show()
```

### Example: Create a figure with 3 subplots (arranged vertically)


```python
fig = plt.figure(figsize=(4, 4))  # 8" wide, 12" tall  

#Line Plot 
ax1 = fig.add_subplot(1, 1, 1)  # 3 rows, 1 column, position 1

#data
random_walk = np.random.standard_normal(50).cumsum()


ax1.plot(random_walk, color="black", linestyle="dashed")

```

```python
# Histogram (Middle) 
fig = plt.figure(figsize=(8, 12))  # 8" wide, 12" tall

ax2 = fig.add_subplot(3, 1, 2)  # Position 2
normal_data = np.random.standard_normal(100)

ax2.hist(normal_data, bins=20, color="black", alpha=0.3)

ax2.set_title("2. Distribution Histogram")
ax2.set_ylabel("Frequency")
```

```python
# Scatter Plot
fig = plt.figure(figsize=(8, 12))  # 8" wide, 12" tall

ax3 = fig.add_subplot(3, 1, 3)  # Position 3
x_values = np.arange(30)
y_values = x_values + 3 * np.random.standard_normal(30)
ax3.scatter(x_values, y_values)
ax3.set_title("3. Noisy Linear Relationship")
ax3.set_xlabel("X Values")
ax3.set_ylabel("Y Values")

plt.tight_layout()  # Prevents label overlapping
plt.show()
```

3- Figures

```python
fig = plt.figure(figsize=(8, 12))  # 8 inches wide, 12 inches tall

ax1 = fig.add_subplot(3, 1, 1)  # (3 rows, 1 column, position 1)
ax1.plot([1, 2, 3, 4], [1, 4, 9, 16], 'r-')  # Red line
ax1.set_title("Quadratic Growth")
ax1.set_ylabel("Y Values")

ax2 = fig.add_subplot(3, 1, 2)  # (3 rows, 1 column, position 2)
ax2.bar(['A', 'B', 'C'], [3, 7, 2], color='blue')  # Blue bars
ax2.set_title("Category Comparison")
ax2.set_ylabel("Counts")

ax3 = fig.add_subplot(3, 1, 3)  # (3 rows, 1 column, position 3)
ax3.scatter(np.random.rand(20), np.random.rand(20), color='green')  # Green dots
ax3.set_title("Random Scatter Plot")
ax3.set_xlabel("X Values")
ax3.set_ylabel("Y Values")

# Final adjustments
plt.tight_layout()  # Automatically spaces subplots nicely
plt.show()
```

## Colors, Markers, and Line Styles

### Colors


```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)
```

```python
y
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Different color specifications
plt.figure(figsize=(10,5))
plt.plot(x, y, color='b')         # Named color;  
plt.plot(x, np.sin(x-1), color='g')          # Short color code (g=green)
plt.plot(x, np.sin(x-2), color='0.75')       # Grayscale (0=black, 1=white)
plt.plot(x, np.sin(x-3), color='#FF0000')    # Hex code (red)
plt.plot(x, np.sin(x-4), color=(1,0,1))      # RGB tuple (purple)
plt.plot(x, np.sin(x-5), color='chartreuse')  # CSS color name
plt.title("Different Color Specifications")
plt.show()# HTML color name
```

```python

Common color codes:

'b' = blue
'g' = green
'r' = red
'c' = cyan
'm' = magenta
'y' = yellow
'k' = black
'w' = white
```

## Line Styles

Control how your lines appear:

```python

# Create sample data
x = np.linspace(0, 10, 100)

# Create figure and plot
plt.figure(figsize=(10, 5))

# Plot with different line styles
plt.plot(x, np.sin(x), linestyle='-', marker='o',label='solid')       # or '-'
plt.plot(x, np.sin(x-1), linestyle='--', label='dashed')   # or '--'
plt.plot(x, np.sin(x-2), linestyle=':', label='dotted')   # or ':'
plt.plot(x, np.sin(x-3), linestyle='-.', label='dashdot') # or '-.'

# Add title and legend
plt.title("Different Line Styles")
plt.legend()
```

```python
Basic Line Styles

'-' = solid line (default)
'--' = dashed line
':' = dotted line
'-.' = dash-dot line
'None' or ' ' = no line
```

###  Markers

Add symbols to your data points:

```python
plt.figure(figsize=(10,5))
plt.plot(x[:15], np.sin(x[:15]), marker='o')      # Circle
```

```python
plt.plot(x[:15], np.sin(x[:15]-1), marker='v')    # Triangle down
plt.plot(x[:15], np.sin(x[:15]-2), marker='^')    # Triangle up
plt.plot(x[:15], np.sin(x[:15]-3), marker='s')    # Square
plt.plot(x[:15], np.sin(x[:15]-4), marker='*')    # Star
plt.plot(x[:15], np.sin(x[:15]-5), marker='+')    # Plus
plt.title("Different Marker Styles")
plt.legend(['o: circle', 'v: triangle down', '^: triangle up', 
            's: square', '*: star', '+: plus'])
plt.show()
```

```python
Markers

'.' = point
',' = pixel
'o' = circle
'v' = triangle_down
'^' = triangle_up
'<' = triangle_left
'>' = triangle_right
'1' = tri_down
'2' = tri_up
'3' = tri_left
'4' = tri_right
'8' = octagon
's' = square
'p' = pentagon
'P' = plus_filled
'*' = star
'h' = hexagon1
'H' = hexagon2
'+' = plus
'x' = x
'X' = x_filled
'D' = diamond
'd' = thin_diamond
'|' = vline
'_' = hline
```

#### Combining All Three

You can combine color, line style, and markers in a compact format:



```python
plt.figure(figsize=(10,5))

plt.plot(x, np.sin(x), 
         color='red', 
         marker='o', 
         linestyle='-')
##Order is always: [color][marker][line]
```

```python
plt.figure(figsize=(10,5))
plt.plot(x, np.sin(x), 'r-o')     #Order is always: [color][marker][line]
```

```python
plt.figure(figsize=(10,5))
plt.plot(x, np.sin(x), 'r-o') 
plt.plot(x, np.cos(x), 'g--s')     # Green dashed line with square markers
plt.plot(x, -np.sin(x), 'b:.')     # Blue dotted line with point markers
plt.title("Combining Color, Line Style, and Markers")
plt.legend(['r-o: red solid circles', 'g--s: green dashed squares', 
            'b:.: blue dotted points'])
```

### Saving Plot

- You can save the active figure to file using the figure object’s `savefig` instance method. 
  
- For example, to save an SVG version of a figure, you need only type: `fig.savefig("figpath.svg")`

- The file type is inferred from the file extension. So if you used .pdf instead, you would get a PDF. 



```python

# Create sample data
x = np.linspace(0, 10, 100)

# Create figure and plot
plt.figure(figsize=(10, 5))

# Plot with different line styles
plt.plot(x, np.sin(x), linestyle='solid', label='solid')       # or '-'
plt.plot(x, np.sin(x-1), linestyle='dashed', label='dashed')   # or '--'
plt.plot(x, np.sin(x-2), linestyle='dotted', label='dotted')   # or ':'
plt.plot(x, np.sin(x-3), linestyle='dashdot', label='dashdot') # or '-.'

# Add title and legend
plt.title("Different Line Styles")
plt.legend()

# # Save the figure in PDF and SVG formats
plt.savefig("line_styles_plot.pdf")
plt.savefig("line_styles_plot.svg")
plt.savefig("line_styles_plot.png")
```

# Viz Quantities

- Visualizing Quantities – Comparing magnitudes. **Magnitude** refers to the *size* or *absolute value* of a quantitative measurement. In visualization, it represents:
  
  - How large or small a value is


- **When to Use:**

  - Comparing discrete counts (sales, votes, etc.)
  




- **Data Types:**
  
  - X-axis: Categorical (nominal/ordinal)
   
  - Y-axis: Numerical (discrete/continuous)

### Real-World Examples

- Election results by political party

- Monthly revenue across departments

- Population by country


### A. Bar Charts


**Best for:** Comparing 3+ categories  

**Rules:**

- Sort bars by value (descending) unless data is ordinal

- Always start y-axis at 0

- Use consistent colors


```python
import matplotlib.pyplot as plt

products = ['Laptops', 'Phones', 'Tablets']
sales = [1200, 950, 300]
```

```python
plt.bar(products, sales, color='skyblue')
plt.title('Q3 Sales by Product')
plt.xlabel('Product Category')
plt.ylabel('Units Sold (thousands)')
plt.show()
```

### A. Column Charts

- Best for: Time-based comparisons

- Pro Tip: Use horizontal bars for long category names







```python
months = ['January', 'February', 'March']
revenue = [50000, 75000, 60000]

plt.barh(months, revenue, color='salmon')
plt.title('Q1 Revenue')
plt.xlabel('Revenue ($)')
plt.show()
```

```python
### Dot Plots

- Best for: Precise value comparison with few data points

- Demo Code:
```

```python
countries = ['USA', 'Canada', 'Mexico']
gdp = [21, 1.6, 1.2]  # in trillions

plt.plot(gdp, countries, 'o', markersize=10)
plt.title('GDP Comparison')
plt.xlabel('GDP ($T)')
plt.grid(axis='x', linestyle='--')
plt.show()
```
